```python
#@title Imports
from diffusers import DiffusionPipeline
from riffusion.audio import wav_bytes_from_spectrogram_image
from io import BytesIO
from IPython.display import Audio

pipeline = DiffusionPipeline.from_pretrained("riffusion/riffusion-model-v1")
pipeline = pipeline.to("cuda")
```

```python
asd
```

```python
#@title Run a Gradio demo
import gradio as gr

def predict(prompt, negative_prompt):
    spec = pipeline(
        prompt,
        negative_prompt=negative_prompt,
        width=768,
    ).images[0]
    wav = wav_bytes_from_spectrogram_image(spec)
    with open("output.wav", "wb") as f:
        f.write(wav[0].getbuffer())
    return 'output.wav'

gr.Interface(
    predict,
    inputs=["text", "text"],
    outputs=gr.outputs.Audio(type='filepath'),
    title="Riffusion",
).launch(share=True, debug=True)
```
